```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from google.colab import files
import joblib
```

**Data Loading**

```python
train_files = [
    "DATA_MEO_Train.csv",
    "DATA_MEO_Train2.csv"
]

test_files = [
    "DATA_MEO_Test.csv",
    "DATA_MEO_Test2.csv"
]
```

```python
train_df_1 = pd.read_csv(train_files[0])
train_df_2 = pd.read_csv(train_files[1])

train_df_1.columns = train_df_1.columns.str.replace(" ", "")
train_df_2.columns = train_df_2.columns.str.replace(" ", "")

print("Train shape:", train_df_1.shape)
print("Test shape:", train_df_2.shape)
print("Columns:", train_df_1.columns.tolist())
print("Columns:", train_df_2.columns.tolist())
```

```python
train_df = pd.concat([train_df_1, train_df_2], axis=0)
print("Train shape:", train_df.shape)
```

```python
train_df.head()
train_df.tail()
```

```python
test_df_1 = pd.read_csv(test_files[0])
test_df_2 = pd.read_csv(test_files[1])

test_df_1.columns = test_df_1.columns.str.replace(" ", "")
test_df_2.columns = test_df_2.columns.str.replace(" ", "")

print("Train shape:", test_df_1.shape)
print("Test shape:", test_df_2.shape)
```

```python
test_df = pd.concat([test_df_1, test_df_2], axis=0)
print("Test shape:", test_df.shape)
```

```python
test_df.head()
```

**Data Cleaning**

```python
train_df["utc_time"] = pd.to_datetime(train_df["utc_time"])
test_df["utc_time"]  = pd.to_datetime(test_df["utc_time"])

train_df = train_df.sort_values("utc_time").reset_index(drop=True)
test_df  = test_df.sort_values("utc_time").reset_index(drop=True)
```

```python
train_df.head()
```

```python
train_df.isna().sum()
```

```python
FEATURE_COLS = ["x_error(m)", "y_error(m)", "z_error(m)", "satclockerror(m)"]
TARGET_COL = "satclockerror(m)"

scaler = MinMaxScaler()

train_df[FEATURE_COLS] = scaler.fit_transform(train_df[FEATURE_COLS])
test_df[FEATURE_COLS]  = scaler.transform(test_df[FEATURE_COLS])
```

```python
train_df.head()
```

**Exporting the Datasets**

```python
train_df.to_csv("train_df.csv", index=False)
files.download("train_df.csv")
test_df.to_csv("test_df.csv", index=False)
files.download("test_df.csv")
```

**Exporting Scaler Model**

```python
joblib.dump(scaler, "scaler_model.pkl")
```
